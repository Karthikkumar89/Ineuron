# 9. how many person are not a regular person with respect to activity try to find out those

```python
df['is Healthy']=df['Calories'].apply(Healt_status)
```

```python
df.head()
```

```python
dfa=pd.DataFrame(df.groupby('Id')['Health status'].count())
```

```python
dfa['cnt_healthy']=pd.DataFrame(df[df['is Healthy']=='Healthy'].groupby('Id')['is Healthy'].count())
```

```python
dfa['cnt_unhealthy']=df[df['is Healthy']=='Un-healthy'].groupby('Id')['is Healthy'].count()
```

```python
dfa['%activity']=round((dfa['cnt_healthy']/dfa['Health status'])*100)
```

```python
#regular >80% activity
dfa['is Regular']=pd.DataFrame(dfa['%activity'].apply(lambda x: 'Y' if x>80 else 'N'))
```

```python
dfa.head()
```

```python
print('Number of persons who are not regular = {}'.format(dfa[dfa['is Regular']=='N']['is Regular'].count()))
```

```python
print('Number of persons who are not regular = {}'.format(len(dfa[dfa['is Regular']=='N'])))
```
